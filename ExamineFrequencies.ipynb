```python
def preview():
    plt.figure(figsize=(14, 8))
    for ii in range(5):
        plt.subplot(3, 5, 10+ii+1)
        plt.imshow(some[ii])
        plt.imshow(mask[ii], alpha=0.5)
        plt.subplot(3, 5, 5+ii+1)
        plt.imshow(some[ii])
        plt.subplot(3, 5, ii+1)
        plt.imshow(__some[ii])
    plt.show()
    plt.close()
    
preview()
```
